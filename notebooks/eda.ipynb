# Key EDA Insights

1. Transaction amounts exhibit a highly right-skewed distribution, indicating the presence of a small number of very large transactions alongside many low-value transactions.

2. Several numerical features contain significant outliers, which may require scaling or transformation during feature engineering.

3. Customer activity is concentrated within a limited number of product categories and transaction channels, suggesting strong behavioral patterns that may help predict risk.

4. Certain numerical variables show strong correlations, indicating potential redundancy and opportunities for feature selection.

5. Transaction timestamps contain valuable behavioral information that can be transformed into time-based features such as hour, day, month, and weekday during feature engineering.
